```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df=pd.read_excel("Amazon_scrapping.xlsx")
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.head()
```

```python
column= ['seller/id', 'title', 'price/value']
```

```python
df[column].isnull().values.any()
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
def get_imp_features(data):
    imp_features = []
    for i in range(0, df.shape[0]):
        imp_features.append("{} {} {}".format(data['seller/id'][i], data['title'][i], data['price/value'][i]))
    return imp_features
```

```python
df['imp_features']=get_imp_features(df)
df.head(2)
```

```python
#text to matrix
cm=CountVectorizer().fit_transform(df['imp_features'])
```

```python
#get cosin similarity
cs=cosine_similarity(cm)
print(cs)
```

```python
cs.shape
```

```python
df['product_id']=range(1,len(df) + 1)
```

```python
title='Apple TV Siri Remote (2nd Generation)'
movies_id=df[df.title == title]['product_id'].values[0]
```

```python
df.head(3)
```

```python
scores=list(enumerate(cs[movies_id]))
```

```python
sorted_scores=sorted(scores,key=lambda x:x[1],reverse=True)
sorted_scores=sorted_scores[1:]
```

```python
print(sorted_scores)
```

```python
j=0
print('top 5 similar products')
for item in sorted_scores:
  product_title=df[df.product_id == item[0]]['title'].values[0]
  print(j+1,product_title)
  j=j+1
  if j>5:
    break
```
